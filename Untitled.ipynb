```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from collections import Counter
from wordcloud import STOPWORDS
```

```python
pd.set_option('display.max_rows', 200)
```

```python
df_train = pd.read_csv('train.csv', dtype={'id': np.int16, 'target': np.int8})
df_test = pd.read_csv('test.csv', dtype={'id': np.int16})
```

```python
print('Training Set Shape = {}'.format(df_train.shape))
print('Training Set Memory Usage = {:.2f} MB'.format(df_train.memory_usage().sum() / 1024**2))
print('Test Set Shape = {}'.format(df_test.shape))
print('Test Set Memory Usage = {:.2f} MB'.format(df_test.memory_usage().sum() / 1024**2))
```

```python
missing_cols = ['keyword', 'location']
df_sum = df_train[missing_cols].isnull().sum()
df_sumtest = df_test[missing_cols].isnull().sum()
fig, axes = plt.subplots(ncols=2, figsize=(17, 4), dpi=100)
sns.barplot(x=df_sum.values, y=df_sum.index, ax=axes[0], palette="muted")
sns.barplot(x=df_sumtest.values, y=df_sum.index, ax=axes[1], palette="muted")
axes[0].set_title('Training set')
axes[1].set_title('Test set')
plt.show()
###############################################################


df_sum = df_train[missing_cols].isnull().sum()
df_sum1 = df_train['keyword'].size
df_sumtest = df_test[missing_cols].isnull().sum()
print('Percentage of missing values in training set:\n', df_sum/df_sum1*100)
print('It seems that training and test set seems to have the same amount of missing data. \n Probably taken from the same distribution')
```

```python
print('number of unique keywords in training set: ' , df_train['keyword'].value_counts().size)
print('number of unique locations in training set: ' , df_train['location'].value_counts().size)
print('number of unique keywords in test set: ' , df_test['keyword'].value_counts().size)
print('number of unique locations in test set: ' , df_test['location'].value_counts().size)
print('===================================================================================')
print('number of train samples: ' , df_train.size)
print('number of test samples: ' , df_test.size)
print('===================================================================================')
print('number of missing location in training: ' , df_train['location'].isnull().sum())
print('number of missing location in training: ' , df_train['keyword'].isnull().sum())
print('number of missing location in training: ' , df_test['location'].isnull().sum())
print('number of missing location in training: ' , df_test['keyword'].isnull().sum())
```

```python
"""
TODO

word_count number of words in text
unique_word_count number of unique words in text
stop_word_count number of stop words in text
url_count number of urls in text
mean_word_length average character count in words
char_count number of characters in text
punctuation_count number of punctuations in text
hashtag_count number of hashtags (#) in text
mention_count number of mentions (@) in text

"""
```

```python
df_train['word_count'] = df_train['text'].apply(lambda x: len(str(x).split()))
df_train['unique_word_count'] = df_train['text'].apply(lambda x: len(set(str(x).split())))
df_train['stopwords_count'] = df_train['text'].apply(lambda x: len([a for a in str(x).lower().split() if a in STOPWORDS] ))
df_train['url_word_count'] = df_train['text'].apply(lambda x: len([a for a in str(x).lower().split() if a in 'http://' or a in 'https://']))
df_train['mean_word_length'] = df_train['text'].apply(lambda x: sum([len(w) for w in str(x).lower().split()])/len(str(x).lower().split()))
df_train['char_count'] = df_train['text'].apply(lambda x: sum([len(w) for w in str(x).lower().split()]))
df_train['punct_count'] = df_train['text'].apply(lambda x: len([w for w in str(x) if w in "."]))
df_train['hash_count'] = df_train['text'].apply(lambda x: len([w for w in str(x) if w in "#"]))
df_train['mention_count'] = df_train['text'].apply(lambda x: len([w for w in str(x) if w in "@"]))

df_train[df_train.mention_count > 1]
#df_train[df_train['text'].str.contains('http://|https://')]
```

```python
df_test['word_count'] = df_test['text'].apply(lambda x: len(str(x).split()))
df_test['unique_word_count'] = df_test['text'].apply(lambda x: len(set(str(x).split())))
df_test['stopwords_count'] = df_test['text'].apply(lambda x: len([a for a in str(x).lower().split() if a in STOPWORDS] ))
df_test['url_word_count'] = df_test['text'].apply(lambda x: len([a for a in str(x).lower().split() if a in 'http://' or a in 'https://']))
df_test['mean_word_length'] = df_test['text'].apply(lambda x: sum([len(w) for w in str(x).lower().split()])/len(str(x).lower().split()))
df_test['char_count'] = df_test['text'].apply(lambda x: sum([len(w) for w in str(x).lower().split()]))
df_test['punct_count'] = df_test['text'].apply(lambda x: len([w for w in str(x) if w in "."]))
df_test['hash_count'] = df_test['text'].apply(lambda x: len([w for w in str(x) if w in "#"]))
df_test['mention_count'] = df_test['text'].apply(lambda x: len([w for w in str(x) if w in "@"]))
```
